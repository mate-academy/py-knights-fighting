```python
class Knight:
    def __init__(
        self,
        name: str,
        power: int,
        hp: int,
        armour: list,
        weapon: dict,
        potion: dict
    ) -> None:
        self.name = name
        self.power = power
        self.protection = 0
        self.hp = hp
        self.armour = armour
        self.weapon = weapon
        self.potion = potion

    def apply_armour(self) -> None:
        for element in self.armour:
            self.protection += element["protection"]

    def apply_weapon(self) -> None:
        self.power += self.weapon["power"]


artur = Knight("Artur", 10, 80, [],{"name": "Metal Sword", "power": 50,},{"name": "Berserk", "effect": {"power": +15,"hp": -5,"protection": +10,}})
```
